Task 1: Analyzing Taylor Swift's Albums

1.   Group by album and calculate the average song length and the total number of songs

2.   Merge with taylor_albums



```python
df1 = (
taylor_album_songs
 .groupby("album_name", as_index=False)
 .agg(avg_song_length=('duration_ms', 'mean'),
      total_songs=('track_number', 'count'))
 .sort_values('total_songs', ascending=False)
)

df1
```

```python
pd.merge(df1, taylor_albums)
```

Task 2: Analyzing Song Features Across Albums

*   Merge the taylor_all_songs dataset with the taylor_albums dataset to include album information for each song. Group by album and calculate the average values of various song features like danceability, energy, valence, etc.




```python
df2 = (
taylor_all_songs.merge(taylor_albums, on='album_name')
 .groupby('album_name', as_index=False)
 .agg(avg_danceability=('danceability', 'mean'),
      avg_energy=('energy', 'mean'),
      avg_valence=('valence', 'mean'))
 .sort_values('album_name')
)

df2
```
