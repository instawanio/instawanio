Task 1: Calculate the average departure delay and median arrival delay for each carrier.

```python
Task_1 = (
    flights
    .groupby("carrier")
    .agg({"dep_delay":"mean", "arr_delay": "median"})
    .reset_index()
)
```

```python
Task_1
```

Task 2: Find the total number of flights and maximum distance flown for each origin airport.

```python
Task_2 = (
    flights
    .groupby("origin")
    .agg({"flight":"count", "distance":"max"})
    .reset_index()
)
```

```python
Task_2
```

Task 3: Calculate the average departure delay and total distance flown for each carrier operating flights from JFK.

```python
Task_3 = (
    flights
    .loc[(flights["origin"] =="JFK")]
    .groupby(["carrier"])
    .agg({"dep_delay":"mean", "distance":"count"})
    .reset_index()
)
```

```python
Task_3
```

Task 4: Find the median air time and average arrival delay for each destination airport.

```python
Task_4 = (
    flights
    .groupby("dest")
    .agg({"air_time":"median", "arr_delay":"mean"})
    .reset_index()
)
```

```python
Task_4
```

Task 5: Calculate the total number of flights and maximum departure delay for each month.

```python
Task_5 = (
    flights
    .groupby("month")
    .agg({"flight":"count", "dep_delay":"max"})
    .reset_index()
)
```
